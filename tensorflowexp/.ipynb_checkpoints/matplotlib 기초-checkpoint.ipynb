```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib nbagg
import numpy as np
import pandas as pd
```

```python
s = pd.Series(np.random.randn(10).cumsum(),index = np.arrange(0,100,10))
```

```python
s.plot()
```

```python
df = pd.DataFrame(np.random.randn(10,4).cumsum(axis = 0),columns = ["A","B","C","D"],
                  index = np.arange(0,100,10))
```

```python
df.plot()
```

```python
s2 = pd.Series(np.random.rand(16),index=list("abcdefghigklmnop").reverse())
```

```python
s2.plot(kind = 'bar')
```

```python
s2.plot(kind = 'barh')
```

```python
df2 = pd.DataFrame(np.random.rand(6,4),index = ["One","Two","Three","Four","Five","Six"],
                  columns = pd.Index(["A","B","C","D"],name = "Genus"))
```

```python
df2.plot(kind = 'bar',stacked = True)
```

```python
x1 = np.random.normal(1,1,size = (100,1))
x2 = np.random.normal(-2,4,size = (100,1))
X = np.concatenate((x1,x2),axis = 1)
```

```python
df3 = pd.DataFrame(X,columns = ["x1","x2"])
```

```python
plt.scatter(df3['x1'],df3['x2'])
```
