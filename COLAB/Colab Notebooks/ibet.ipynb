```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
diabetes.data.shape,diabetes.target.shape
```

```python
plt.scatter(diabetes.data[:,2],diabetes.target)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x = diabetes.data[:,2]
y = diabetes.target

w = 1.0
b = 1.0

for i in range(100):
  for x_i,y_i in zip(x,y):
    y_hat = x_i * w +b
    err = y_i - y_hat
    w_rate = x_i;
    w =  w + w_rate * err
    b = b + 1 * err
print(w,b);
```

```python
plt.scatter(x,y)
pt1 = (-0.1,-0.1*w+b)
pt2 = (0.15,0.15*w+b)
plt.plot([pt1[0],pt2[0]],[pt1[1],pt2[1]])
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
class Neuron:
  def __init__(self):
    #초기화 작업
    self.w = 1.0
    self.b = 1.0
  def forpass(self,x):
    y_hat = x * self.w + self.b
    return y_hat
  def backdrop(self,x,err):
    w_grad = x * err
    b_grad = 1 * err
    return w_grad,b_grad
  def fit(self,x,y,epochs = 100):
    for i in range(epochs):
        for x_i,y_i in zip(x,y):
          y_hat = self.forpass(x_i)
          err = -(y_i - y_hat)
          w_grad,b_grad =  self.backdrop(x_i,err)
          self.w -= w_grad
          self.b -= b_grad
  
  
```

```python
neuron = Neuron()
neuron.fit(x,y)
```
