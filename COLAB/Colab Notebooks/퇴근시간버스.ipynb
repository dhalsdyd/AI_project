```python
import numpy as np

def get_day(df2):
  df = df2.copy()
  df['weekday'] = df['date'].dt.dayofweek
  holiday=['2019-09-12', '2019-09-13', '2019-09-14','2019-10-03','2019-10-09']
  df['weekends'] = np.where(df['weekday'] >= 5, 1 ,0)
  df['holiday'] = np.where(df['date'].isin(holiday),1,0)
  return df

```

```python
from sklearn.preprocessing import LabelEncoder

def df_copy(train_df, test_df):
    train = train_df.copy(); test = test_df.copy()
    return train, test
def base_preprocessing(train_df, test_df):
    train, test = df_copy(train_df, test_df)

    train['bus_route_id'] = train['bus_route_id'].apply(lambda x: str(x)[:-4]).astype(int)
    test['bus_route_id'] = test['bus_route_id'].apply(lambda x: str(x)[:-4]).astype(int)
    train['station_name2'] = train['station_name'].apply(lambda x: str(x)[:2])
    test['station_name2'] = test['station_name'].apply(lambda x: str(x)[:2])
    train['station_name'] = train['station_name'].apply(lambda x: x.replace(' ', ''))
    test['station_name'] = test['station_name'].apply(lambda x: x.replace(' ', ''))

    le = LabelEncoder().fit(pd.concat([train['station_name'], test['station_name']]))
    le2 = LabelEncoder().fit(pd.concat([train['station_name2'], test['station_name2']]))
    for df in [train, test]:
        df['day'] = pd.to_datetime(df['date']).dt.day
        df['week'] = pd.to_datetime(df['date']).dt.week
        df['weekday'] = pd.to_datetime(df['date']).dt.weekday
        df['station_name'] = le.transform(df['station_name'])
        df['station_name2'] = le2.transform(df['station_name2'])

        df['6~8_ride'] = df[['6~7_ride','7~8_ride']].sum(1)
        df['6~9_ride'] = df[['6~7_ride','7~8_ride','8~9_ride']].sum(1)
        df['6~10_ride'] = df[['6~7_ride','7~8_ride','8~9_ride', '9~10_ride']].sum(1)
        df['6~8_takeoff'] = df[['6~7_takeoff','7~8_takeoff']].sum(1)
        df['6~9_takeoff'] = df[['6~7_takeoff','7~8_takeoff','8~9_takeoff']].sum(1)
        df['6~10_takeoff'] = df[['6~7_takeoff','7~8_takeoff','8~9_takeoff', '9~10_takeoff']].sum(1)
    print(test['day'])
    return train,test
```

```python
def encode_AG(main_columns, uids, aggregations=['mean'], train_df=train, test_df=test, fillna=True, usena=False):
    for main_column in main_columns:  
        for col in uids:
            for agg_type in aggregations:
                new_col_name = main_column+'_'+col+'_'+agg_type
                temp_df = pd.concat([train_df[[col, main_column]], test_df[[col,main_column]]])
                temp_df = temp_df.groupby([col])[main_column].agg([agg_type]).reset_index().rename(columns={agg_type: new_col_name})
                temp_df.index = list(temp_df[col])
                temp_df = temp_df[new_col_name].to_dict()   
                train_df[new_col_name] = train_df[col].map(temp_df).astype('float32')
                test_df[new_col_name]  = test_df[col].map(temp_df).astype('float32')
              
                
                print("'"+new_col_name+"'",', ',end='')
```

```python
import gc

# LABEL ENCODE
def encode_LE(col,x_train=train,x_test=train,verbose=True):
    df_comb = pd.concat([x_train[col],x_test[col]],axis=0)
    df_comb,_ = df_comb.factorize(sort=True)
    nm = col
    if df_comb.max()>32000: 
        x_train[nm] = df_comb[:len(x_train)].astype('int32')
        test[nm] = df_comb[len(x_train):].astype('int32')
    else:
        x_train[nm] = df_comb[:len(x_train)].astype('int16')
        x_test[nm] = df_comb[len(x_train):].astype('int16')
    del df_comb; x=gc.collect()
    if verbose: print(nm,', ',end='')
def encode_CB(col1,col2,df1=train,df2=train):
    nm = col1+'_'+col2
    df1[nm] = df1[col1].astype(str)+'_'+df1[col2].astype(str)
    df2[nm] = df2[col1].astype(str)+'_'+df2[col2].astype(str)
    encode_LE(nm,x_train=df1, x_test=df2, verbose=False)
    print(nm,', ',end='')
```
