## 2. 데이터 훈련 시켜 예측값 생성하기 



```python
import os
import pandas as pd
from datetime import datetime 
from sklearn.linear_model import LinearRegression
# 데이터 경로 설정 
INIT_PATH = os.path.join(current_path, '2021_data_creator_camp', 'data')
SAVE_PATH = os.path.join(current_path, '2021_data_creator_camp', 'save_csv')
os.makedirs(SAVE_PATH, exist_ok=True) 

train_data = pd.read_csv(INIT_PATH+'/sample_train.csv')
test_data = pd.read_csv(INIT_PATH+'/sample_test.csv')
```

```python
# 훈련/평가 데이터 나누기 
X_train = train_data.drop(['id','species'],axis=1)
y_train = train_data['species']

X_test = test_data.drop(['id'],axis=1)
y_test = test_data['id']
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test) # 예측값 
```

```python
today = datetime.now().strftime("%Y%m%d_%H%M%S")
name = 'sample_submission'
csv_name = os.path.join(SAVE_PATH, f'{name}_{today}.csv')

# 결과 파일 저장 
submmision = pd.DataFrame(data=y_pred, columns=['species'])
submmision['id'] = test_data['id']
submmision.to_csv(csv_name,index=False)

display(submmision.head())
# print(f'|INFO| DATE: {today}')
print(f'|INFO| 제출 파일 저장 완료: {csv_name}')
    
```
